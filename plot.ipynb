```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("resultados.csv")

N = df["N"]
insertion = df["Insertion(microsseg)"]*10^6
merge = df["Merge(microsseg)"]*10^6

# curvas teóricas (normalizadas)
n2 = N**2
nlogn = N * np.log2(N)

n2 = n2 / max(n2) * max(insertion)

nlogn = nlogn / max(nlogn) * max(merge)


# plot
plt.figure()

plt.plot(N, insertion, label="Insertion Sort (real)")
plt.plot(N, merge, label="Merge Sort (real)")

plt.plot(N, n2, '--', label="O(n²) (teórico)")
plt.plot(N, nlogn, '--', label="O(n log n) (teórico)")

plt.xlabel("N (número de jogadores)")
plt.ylabel("Tempo (segundos)")
plt.title("Comparação de desempenho")
plt.legend()
plt.grid()

plt.show()
```
